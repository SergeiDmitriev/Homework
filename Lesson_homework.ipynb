```python
list_num1 = []

for num in range(0, 1000):
    if num % 3 == 0 and num % 9 == 0:
        list_num1.append(num)
        
print(list_num1)
print("\n")

list_num2 = [num for num in range(0, 1000) if num % 3 == 0 and num % 9 == 0]
print(list_num2)
```

```python
comparison1 = {}

for num in range(65, 122):
    comparison1[num] = chr(num)
        
print(comparison1)
print("\n")

comparison2 = {num: chr(num) for num in range(65, 122)}
print(comparison2)
```

```python
import itertools
x = [1, 1, 2]

results = list(itertools.permutations(x))

print(results)

#Только уникальные
print("\n")

results1 = list(set(results))

print(results1)
```
